First up the `while` loop:

```python
while (minimum < maximum): # Keep guessing until there is some uncertainty in the guess
```

The generic format is `while (condition is True): #Do something`

The implicit assumption is that once the `condition is False` the block of `while` loop stops running. So in the above code when the condition `minimum < maximum` is false, i.e. `minimum >= maximum` then the code executes the next lines of code after the `while` code block is done, i.e. the following lines will then be run

```python
# At this point minimum=maximum so we know the actual correct value!
print("--------------------------------")
print("The correct word limit is ="+str(minimum))
```
